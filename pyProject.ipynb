L'objectif de ce projet est de proposer un programme en Python qui génère automatiquement de matricules des étudiants de l'IMSP à partir de leurs nom et prénom. Le matricule demenadé comporte trois parties dont les deux premières sont espectivement les trois premières consonnes du nom, les trois premières lettres du prénom.La troisième partie est un nombre à trois chiffres entre 001 et 999 qui sert à gérer le cas des des matricules dont les deux premiers champs sont identiques.

Pour réaliser ce projet, on demande à l'utilisateur d'entrer les nom et prénom de l'étudiant, on vérifie avec les expressions régulières la présence de caractères spéciaux.Les prénoms contenant des lettres accentuées ont été pris en compte en remplaçant dans le matricule ces lettrespar les lettres majuscule de base ( par exemple â est remplacé par A)
    Pour le troisième champ du matricule, nous avons décidé de l'initialiser à 001 pour tous les étudiants car une fois attribué, un matricule ne peut être changé et donc au cas où deux étudiants auraient les deux premiers champs de leurs matricules identiques le premier étudiant enregistré aurait déjà eu 001 et le second 002
    L'analyse du projet nous a amené à enregistrer les matricules générés dans un fichier nommé matricule.txt
    Ce fichier contient en plus le nom et le prénom correspondant aux différents matricules

```python
import re
#importation du module re pour utiliser les expressions régulières
def isConsonent(letter):
    """Cette fonction prend en paramètre un caractère et retourne True s'il est une consonne(en majuscule) et False sinon"""
    
    if re.search("[A-Z]",letter):
    #si letter est une lettre majuscule
        
        if letter in {'A','E','I','O','U','Y'} :
        #Si letter est une voyelle majuscule
            return False
        else :
            return True
    return False #On retourne False aussi quand letter n'est pas une lettre alphabétique(majuscule)

def transformToVowel(char):
    """Cette fonction  transforme  une lettre majuscule accentuée  prise en paramètre en son lettre  non accentuéekii associé"""
    dicte = {'A':['À','Á','Â', 'Ä','Å','Æ'], 'E':['È','É','Ê','Ë'], 'I':['Ì','Í','Î','Ï'],
            'O':['Ò','Ó','Ô','Õ','Ö'],'U':['Ù','Ú','Û','Ü'],'Y':['Ý']}
    var =list()
    for i in dicte.keys():
        var.append(i)
    if isConsonent(char)==False:
        if(char not in var):
            for key in var :
                for c in dicte[key] :
                    if c == char :
                        return key
        else :
            return char 
    else :
        return char
        
                



def generateMatricule(firstname,lastname):
    """La fonction prend en paramètre le prénom et le nom de l'étudiant, génère et retourne les deux premiers champs du matricule correspondant en se basnt sur le principe décrit"""
    firstnameUpper =firstname.upper()#Contient le mot correspondant en majuscule
    lastnameUpper =lastname.upper()
    cpt=0
    matricule=""
    for letter in lastnameUpper :
        if cpt < 3 :
            if isConsonent(letter) :
                matricule+=letter
                cpt+=1
        else :
            break
    if cpt<3 :
        matricule+=(3-cpt)*"X"
    if len(firstnameUpper)<=3:
        for letter in firstnameUpper[:len(firstnameUpper)]:
            matricule+=transformToVowel(letter)
        matricule+=(3-len(firstnameUpper))*"X"
    else :
        for letter in firstnameUpper[:3]:
            matricule+=transformToVowel(letter)
        
        
    return matricule
            
            
#generateMatricule("Mrcel","Gnansounou")    
                

        
    
        
    
```

```python
def matriculeOccur(matricule):
    """Cette fonction ouvre le fichier matricule en mode écriture et retourne le nombre de matricule(s) ayant leurs deux premiers champs identiques"""
    doc=open("matricule.txt","r")
    '''print(doc.readline(), end='')
    print(doc.readline())'''
    count=1
    for line in doc:
        
        var=doc.readline()[:6]
        if matricule==var :
            count+=1
    doc.close()
    return count
```

```python
def writeMatricule(firstname,lastname) :
    """Cette fonction écrit le matricule généré à partir du nom et du prénom pris en paramètre dans le fichier matricule.txt"""
    matricule=generateMatricule(firstname,lastname)
    var = matriculeOccur(matricule)
    if var <10:
        matricule+=("00"+str(var))
    elif  var <100 :
        matricule+=("0"+str(var))
    else :
        matricule+=str(var)
    doc=open("matricule.txt","a+")
    
    doc.write("\n"+matricule+" : "+firstname+" "+lastname+"\n")
    
    doc.close()

```
